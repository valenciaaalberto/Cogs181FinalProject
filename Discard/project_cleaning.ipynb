```python
import string
import random
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import random
```

```python
folders = open('/Users/albertovalencia/Downloads/tiny-imagenet-200/wnids.txt').readlines()
train_file_path = "/Users/albertovalencia/Downloads/tiny-imagenet-200/train/"
test_file_path = "/Users/albertovalencia/Downloads/tiny-imagenet-200/test/"
train_folders = os.listdir("/Users/albertovalencia/Downloads/tiny-imagenet-200/train/")
train_folders.remove(".DS_Store")
test_folders = [""]
```

```python
def create_set(test_or_train_path,folders):
    test_or_train_set = []
    for folder in folders:
        folder_path = test_or_train_path + folder.strip() + "/images/"
        file_paths = os.listdir(folder_path)
        for file_path in file_paths:
            current_image_path = folder_path + file_path
            current_image = open(current_image_path)
            current_image = cv2.imread(current_image_path)
            test_or_train_set.append(np.array(current_image))
    return np.array(test_or_train_set)
```

```python
trainset = create_set(train_file_path,train_folders)
testset = create_set(test_file_path,test_folders)
```

```python
import string
import random
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import random
```

```python
train = []
folders = open('/Users/albertovalencia/Downloads/tiny-imagenet-200/wnids.txt').readlines()
for folder in folders:
    folder_path = "/Users/albertovalencia/Downloads/tiny-imagenet-200/train/" + folder.strip() + "/images/"
    file_paths = os.listdir(folder_path)
    for file_path in file_paths:
        current_image_path = folder_path + file_path
        current_image = open(current_image_path)
        current_image = cv2.imread(current_image_path)
        train.append(np.array(current_image))
train = np.array(train)
```

```python
import string
import random
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import random
import deeplake
import torchvision
import torchvision.transforms as transforms
```

```python
trainset = deeplake.load("hub://activeloop/tiny-imagenet-train")
testset = deeplake.load("hub://activeloop/tiny-imagenet-test")
validationset = deeplake.load("hub://activeloop/tiny-imagenet-validation")
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainloader= trainset.pytorch(batch_size = 4, num_workers = 0, 
    transform = {'images': transform, 'labels': None}, shuffle = False)
```

```python
def imshow(img):
    img = img / 2 + 0.5     # Unnormalize.
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
dataiter = iter(trainloader)
images,label = next(dataiter)
```

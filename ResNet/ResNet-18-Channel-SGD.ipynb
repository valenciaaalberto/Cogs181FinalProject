# Resnet 18 with Channel Attention

## Channel Attention

```python
class SEBlock(nn.Module):
    def __init__(self, channel, reduction=16):
        super(SEBlock, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channel, channel // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channel // reduction, channel, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.avg_pool(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y.expand_as(x)
```

```python
class ResidualBlockWithSE(nn.Module):
    def __init__(self, block, channel):
        super(ResidualBlockWithSE, self).__init__()
        self.block = block
        self.se = SEBlock(channel)

    def forward(self, x):
        out = self.block(x)
        out = self.se(out)
        return out
```

```python
import torch
import torch.nn as nn
import torchvision.models as models

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        resnet18 = models.resnet18(pretrained=False)

        # Modify the first convolutional layer to suit CIFAR-100 image size
        resnet18.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
        
        # Modify the fully connected layer to output 100 classes
        num_ftrs = resnet18.fc.in_features
        resnet18.fc = nn.Linear(num_ftrs, 100)
        
        # Replace blocks in layer1 and layer2 with SE blocks
        resnet18.layer1 = nn.Sequential(
            ResidualBlockWithSE(resnet18.layer1[0], 64),
            ResidualBlockWithSE(resnet18.layer1[1], 64)
        )
        resnet18.layer2 = nn.Sequential(
            ResidualBlockWithSE(resnet18.layer2[0], 128),
            ResidualBlockWithSE(resnet18.layer2[1], 128)
        )
        
        # Replace blocks in layer3 with SE blocks
        resnet18.layer3 = nn.Sequential(
            ResidualBlockWithSE(resnet18.layer3[0], 256),
            ResidualBlockWithSE(resnet18.layer3[1], 256)
        )

        # Replace blocks in layer4 with SE blocks
        resnet18.layer4 = nn.Sequential(
            ResidualBlockWithSE(resnet18.layer4[0], 512),
            ResidualBlockWithSE(resnet18.layer4[1], 512)
        )

        self.features = nn.Sequential(*list(resnet18.children())[:-2])  # Remove the last FC layer
        self.fc = resnet18.fc  # You might need to adjust this for your specific task


    def forward(self, x):
        x = self.features(x)
        x = F.adaptive_avg_pool2d(x, (1, 1))
        x = torch.flatten(x, 1)
        x = self.fc(x)
        return x

# Assuming 'device' is defined (e.g., 'cuda' or 'cpu')
net = Net()
net.to(device)
```

```python
# We use cross-entropy as loss function.
loss_func = nn.CrossEntropyLoss()  
# We use stochastic gradient descent (SGD) as optimizer.
opt = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
avg_losses = []   # Avg. losses.
epochs = 90       # Total epochs.
print_freq = 1000  # Print frequency.

for epoch in range(epochs):  # Loop over the dataset multiple times.
    running_loss = 0.0       # Initialize running loss.
    for i, data in enumerate(trainloader, 0):
        # Get the inputs.
        inputs, labels = data
        
        # Move the inputs to the specified device.
        inputs, labels = inputs.to(device), labels.to(device)
        
        # Zero the parameter gradients.
        opt.zero_grad()

        # Forward step.
        outputs = net(inputs)
        loss = loss_func(outputs, labels)
        
        # Backward step.
        loss.backward()
        
        # Optimization step (update the parameters).
        opt.step()

        # Print statistics.
        running_loss += loss.item()
        if i % print_freq == print_freq - 1: # Print every several mini-batches.
            avg_loss = running_loss / print_freq
            print('[epoch: {}, i: {:5d}] avg mini-batch loss: {:.3f}'.format(
                epoch, i, avg_loss))
            avg_losses.append(avg_loss)
            running_loss = 0.0

print('Finished Training.')
```

```python
plt.plot(avg_losses)
plt.xlabel('mini-batch index / {}'.format(print_freq))
plt.ylabel('avg. mini-batch loss')
plt.show()
```

```python
# Get test accuracy.
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        images, labels = images.to(device), labels.to(device)
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy of the network on the X test images: %d %%' % (
    100 * correct / total))
```
